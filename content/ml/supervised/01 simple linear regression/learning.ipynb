```python
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import SGDRegressor
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('data/advertising.csv').to_numpy()
```

```python
data
```

```python
X = data[:, 0:3]
y = data[:, 3]
```

$$
\hat{y} = \beta_0 + \beta_1 x
$$

$\beta_0$ - intercept
$\beta_1$ - weight

```python
def f(x):
    return 2.48 + 9.77*x
```

```python
y_pred = f(X.reshape(-1))
```

```python
plt.scatter(X.reshape(-1), y)
plt.scatter(X.reshape(-1), y_pred)
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
plt.scatter(X.reshape(-1), y)
plt.scatter(X.reshape(-1), model.predict(X))
plt.show()
```
