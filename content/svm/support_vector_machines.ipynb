```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
df = pd.DataFrame(data=[
    [1, 1, 0], [1, 2, 0], [-1, 1, 0], [-1, -1, 0], [2, -2, 0], [1, -1, 0],
    [0.5, 0.5, 0], [0.8, 1, 0], [-0.5, 1, 0], [-1.2, -1.1, 0], [1.8, -0.5, 0], [1.4, -1.8, 0],
    [4, 5, 1], [-4, -3, 1], [5, -6, 1], [1, -7, 1], [2, 5, 1], [-1, 5, 1],
    [1, 5, 1], [-5, -2, 1], [2, -6, 1], [1, -7, 1], [-3, 4, 1], [-3, -5, 1],
], columns=['x', 'y', 'label'])

sns.scatterplot(data=df, x='x', y='y', hue='label')
plt.axhline(y=0)
plt.axvline(x=0)
plt.show()
```
