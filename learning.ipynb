```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 4, 5, 3, 2, 7])
y = np.array([1, 5, 3, 2, 4, 6])

X = X.reshape((len(X),1))
X = np.c_[np.ones((len(X), 1)), X]
```

$$
MSE(\beta) = \frac{1}{n}(y^Ty - 2\beta^TX^Ty + \beta^TX^TX\beta)
$$

```python
beta = np.array([1.14, 0.64])
mse = (y.T @ y - 2*beta.T@X.T@y + beta.T@X.T@X@beta) / len(y)
mse
```

$$
\hat{\beta} = (X^TX)^{-1}X^Ty
$$

```python
beta = np.linalg.inv(X.T@X)@X.T@y
beta
```

$$
f(x) = e^x
$$

```python
x = np.linspace(-10, 10, 10001)
f_x = np.exp(x)
plt.plot(x, f_x, label="$y=e^x$")
plt.legend()
plt.show()
```

```python
f_x = np.sin(x)
dydx = np.gradient(f_x, x)
plt.figure(figsize=(7, 3))
plt.plot(x, f_x, label='f(x) = sin(x)')
plt.plot(x, dydx, label="f(x) = sin(x)'")
plt.axhline(y=0, c='black', linewidth=0.5)
plt.axvline(x=0, c='black', linewidth=0.5)
plt.legend()
plt.show()
```

```python
weight = .5
input = 0.5
goal_prediction = 0.8

step_size = 0.001

print(f"{input=}, {goal_prediction=}, {step_size=}")

weights = [weight]
errors = []
for i in range(1101):
    prediction = input * weight
    error = goal_prediction - prediction
    print(f"{weight=:<20} {prediction=:<20} {error=}")
    if error**2 < 0.00001:
        print('BREAK')
        break
    if error > 0:
        weight += step_size
    else:
        weight -= step_size
    weights.append(weight)
    errors.append(error)
else:
    print('COMPLETED WITHOUT BREAK')
```

```python
import matplotlib.pyplot as plt
plt.plot(weights, marker='o')
plt.show()
plt.plot(errors)
plt.show()
```
