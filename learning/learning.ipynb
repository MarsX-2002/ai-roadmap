```python
import numpy as np

X = np.random.random(size=(100,1))
y = 5 + 2*X + 0.5*np.random.random(size=(100, 1))
y = y.ravel()
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(X, y)
```

```python
y_pred1 = lr.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred1, 'r', label='Linear Regression')
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend()
plt.show()
```

```python
from sklearn.svm import SVR

svr = SVR(kernel='linear')
svr.fit(X, y)
```

```python
y_pred2 = svr.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred1, 'r', label='Linear Regression')
plt.plot(X, y_pred2, 'g', label='SVR')
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend()
plt.show()
```

```python
from sklearn.tree import DecisionTreeRegressor

dtr = DecisionTreeRegressor()
dtr.fit(X, y)
```

```python
y_pred3 = dtr.predict(X)
```

```python
plt.scatter(X, y_pred3)
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.show()
```

```python
from sklearn.linear_model import SGDRegressor

sgd = SGDRegressor()
sgd.fit(X, y)
```

```python
y_pred4 = sgd.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred1, 'r', label='Linear Regression')
plt.plot(X, y_pred2, 'g', label='SVR')
plt.plot(X, y_pred4, 'y', label='SGDRegressor')
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import Lasso

lasso = Lasso()
lasso.fit(X, y)
```

```python
y_pred5 = lasso.predict(X)
```

```python
plt.scatter(X, y)
plt.plot(X, y_pred1, 'r', label='Linear Regression')
plt.plot(X, y_pred2, 'g', label='SVR')
plt.plot(X, y_pred4, 'y', label='SGDRegressor')
plt.plot(X, y_pred5, 'k', label='Lasso')
plt.xlabel("$x$")
plt.ylabel("$y$")
plt.legend()
plt.show()
```
