```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from scipy.spatial import Voronoi, voronoi_plot_2d
```

```python
np.random.seed(0)
data = np.random.rand(100, 2)

kmeans = KMeans(n_clusters=5, random_state=0)
kmeans.fit(data)

centroids = kmeans.cluster_centers_

vor = Voronoi(centroids)

fig, ax = plt.subplots(figsize=(6, 6))
voronoi_plot_2d(vor, ax=ax, show_verticles=False, line_colors='orange', line_width=2)
ax.scatter(data[:,0], data[:,1], c=kmeans.labels_, cmap='tab10', s=30, label='Data points')
ax.scatter(centroids[:,0], centroids[:,1], color='red', marker='x', s=200, label='Centroids')

ax.legend(loc='center left', bbox_to_anchor=(1, 0.5))
ax.set_title('Voronoi Tesselation with K-Means Clustering')
plt.tight_layout()
plt.show()
```
